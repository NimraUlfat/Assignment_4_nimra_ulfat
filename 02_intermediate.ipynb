Control Flow for Console

```python
import random

# Problem Statement:
# The High-Low game involves two numbers being generated: one for the player and one for the computer (both between 1 and 100).
# The player guesses if their number is higher or lower than the computer's number.
# If the player's guess is correct, they score a point.
# After a set number of rounds (NUM_ROUNDS), the game ends and the player's score is evaluated.
# The game gives feedback based on how well the player performed.

# Constants
NUM_ROUNDS = 5  # You can adjust this as needed for the number of rounds

# Function to play a round
def play_round(round_number, score):
    # Generate random numbers for the player and the computer
    player_number = random.randint(1, 100)
    computer_number = random.randint(1, 100)

    # Print round details
    print(f"Round {round_number}")
    print(f"Your number is {player_number}")

    # Get user input for their guess
    guess = input(f"Do you think your number is higher or lower than the computer's?: ").lower()

    # Validate input to ensure it is either 'higher' or 'lower'
    while guess not in ['higher', 'lower']:
        guess = input("Please enter either 'higher' or 'lower': ").lower()

    # Check if the player's guess is correct
    if (guess == 'higher' and player_number > computer_number) or (guess == 'lower' and player_number < computer_number):
        print(f"You were right! The computer's number was {computer_number}")
        score += 1  # Increment score if the guess is correct
    else:
        print(f"Aww, that's incorrect. The computer's number was {computer_number}")

    # Print current score after the round
    print(f"Your score is now {score}\n")
    return score

# Function to give feedback at the end of the game
def give_feedback(score, rounds_played):
    # Evaluate performance based on the score
    if score == rounds_played:
        print("Wow! You played perfectly!")  # Perfect game
    elif score >= rounds_played // 2:
        print("Good job, you played really well!")  # Half or more rounds won
    else:
        print("Better luck next time!")  # Less than half rounds won

# Main game function
def play_game():
    print("Welcome to the High-Low Game!")
    print("--------------------------------")

    score = 0  # Initialize score at the start

    # Play multiple rounds
    for round_number in range(1, NUM_ROUNDS + 1):
        score = play_round(round_number, score)

    # End game feedback
    print("Thanks for playing!")
    give_feedback(score, NUM_ROUNDS)

# Run the game
play_game()
```

```python
# Planetary Weight Calculator
# This program calculates a person's weight on different planets based on gravitational differences.
# The user can choose to calculate their weight on Mars specifically or on any other planet.

# Milestone #1: Mars Weight
# Mars has a gravitational pull that is 37.8% of Earth's gravity.
# We define a constant MARS_MULTIPLE to store this value.

MARS_MULTIPLE = 0.378

def mars_weight():
    """Prompts the user for their weight on Earth and calculates the equivalent weight on Mars."""
    earth_weight = float(input("Enter a weight on Earth: "))  # Get the user's weight on Earth
    mars_weight = round(earth_weight * MARS_MULTIPLE, 2)  # Calculate weight on Mars
    print(f"The equivalent on Mars: {mars_weight}")  # Display result

# Milestone #2: Adding in All Planets
# Defining gravitational constants for each planet relative to Earth's gravity
MERCURY_GRAVITY = 0.376
VENUS_GRAVITY = 0.889
MARS_GRAVITY = 0.378
JUPITER_GRAVITY = 2.36
SATURN_GRAVITY = 1.081
URANUS_GRAVITY = 0.815
NEPTUNE_GRAVITY = 1.14

def planetary_weight():
    """Prompts the user for their weight on Earth and a planet name, then calculates equivalent weight."""
    earth_weight = float(input("Enter a weight on Earth: "))  # Get the user's weight on Earth
    planet = input("Enter a planet: ")  # Get the target planet name

    gravity_factors = {
        "Mercury": MERCURY_GRAVITY,
        "Venus": VENUS_GRAVITY,
        "Mars": MARS_GRAVITY,
        "Jupiter": JUPITER_GRAVITY,
        "Saturn": SATURN_GRAVITY,
        "Uranus": URANUS_GRAVITY,
        "Neptune": NEPTUNE_GRAVITY
    }

    if planet in gravity_factors:
        planetary_weight = round(earth_weight * gravity_factors[planet], 2)  # Calculate weight on chosen planet
        print(f"The equivalent weight on {planet}: {planetary_weight}")  # Display result
    else:
        print("Invalid planet name. Please enter a valid planet.")  # Handle invalid input

def main():
    """Main function to ask user whether they want to calculate weight for Mars or another planet."""
    choice = input("Do you want to calculate weight for Mars or another planet? (Mars/Other): ").strip()
    if choice.lower() == "mars":
        mars_weight()
    else:
        planetary_weight()

# Entry point of the program
if __name__ == "__main__":
    main()
```

```python
# Problem #1: List Practice
# This program demonstrates basic list operations, including initialization, modifying, and printing a list.

def main():
    # Create a list called fruit_list that contains the following fruits:
    # 'apple', 'banana', 'orange', 'grape', 'pineapple'.
    fruit_list = ['apple', 'banana', 'orange', 'grape', 'pineapple']

    # Print the length of the list
    print("Length of the list:", len(fruit_list))

    # Add 'mango' at the end of the list
    fruit_list.append('mango')

    # Print the updated list
    print("Updated list:", fruit_list)

# Problem #2: Index Game
# This program allows the user to interactively access, modify, and slice a list.
# The user is given options to perform different operations on a predefined list.

def access_element(lst, index):
    """Returns the element at the specified index or an error message if out of range."""
    try:
        return lst[index]
    except IndexError:
        return "Index out of range."

def modify_element(lst, index, new_value):
    """Replaces the element at the specified index with a new value."""
    try:
        lst[index] = new_value
        return "Element updated successfully."
    except IndexError:
        return "Index out of range."

def slice_list(lst, start, end):
    """Returns a new list containing elements from start index up to (but not including) end index."""
    return lst[start:end]

def index_game():
    """Interactive game that allows the user to access, modify, and slice a list."""
    elements = [10, 20, 30, "apple", "banana"]

    while True:
        print("\nCurrent List:", elements)
        print("Choose an operation:")
        print("1. Access an element")
        print("2. Modify an element")
        print("3. Slice the list")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            index = int(input("Enter the index: "))
            print("Result:", access_element(elements, index))
        elif choice == '2':
            index = int(input("Enter the index: "))
            new_value = input("Enter the new value: ")
            print(modify_element(elements, index, new_value))
        elif choice == '3':
            start = int(input("Enter start index: "))
            end = int(input("Enter end index: "))
            print("Sliced List:", slice_list(elements, start, end))
        elif choice == '4':
            print("Exiting the game.")
            break
        else:
            print("Invalid choice. Please try again.")

# Entry point of the program
if __name__ == "__main__":
    main()
    index_game()
```
